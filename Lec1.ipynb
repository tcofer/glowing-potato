## Jupyter Notebooks



Jupyter Notebooks are a web-based interactive developement environment that combine live code, text, and other media.

#### Modes 

Notebooks have two modes, a `command mode` and `edit mode`. You can see which mode you're in by the color of the cell: <br><br>
<span style="color:green">Green </span> means edit mode. (You can type into the cell.) <br>
<span style="color:blue"> Blue </span> means command mode. (You can edit the notebook as a whole, but you cannot type into individual cells.) <br>

#### Notebook Shortucts

While you can always use the tool-bar above, you'll be much more efficient if you use a couple of shortcuts. The most important ones are:

**`Ctrl+Enter`** runs the current cell.  
**`Shift+Enter`** runs the current cell and jumps to the next cell.   
**`Alt+Enter`** runs the cell and adds a new one below it.

In command mode:

**`h`** shows a help menu with all these commands.  
**`a`** adds a cell before the current cell.  
**`b`** adds a cell after the current cell.  
**`dd`** deletes a cell.  
**`m`** as in **m**arkdown, switches a cell to markdown mode.  
**`y`** as in p**y**thon switches a cell to code.

##### A Note on Additional Libraries

The core python language is extended by standard libraries that provide additional functionality in the form of modules that we can *import* into our python session.

For example, the `math` module provides functions that do the basic mathematical operations as well as provide constants.

In python, you `import` a module.  The functions are then defined in a separate _namespace_&mdash;this is a separate region that defines names and variables, etc. You use the "`.`" operator to access a member of a namespace.

By default, when you type stuff into the python interpreter or here in the Jupyter notebook, or in a script, it is in its own default namespace, and you don't need to prefix any of the variables with a namespace indicator.

```python
import math
```

`math` provides the value of pi

```python
print(math.pi)
```
